# Phase 1 Data Exploration
Inspect pipeline_report.json, sample HR/LR pairs, and patch statistics.


```python
import json
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

report_path = Path('../reports/pipeline_report.json')
report = json.loads(report_path.read_text())
report
```

```python
samples = report.get('samples', {})
for key in ['sample_hr', 'sample_lr', 'sample_lr_upscaled_preview']:
    if key in samples:
        img = Image.open(samples[key])
        plt.figure()
        plt.title(key)
        plt.imshow(img)
        plt.axis('off')
```
